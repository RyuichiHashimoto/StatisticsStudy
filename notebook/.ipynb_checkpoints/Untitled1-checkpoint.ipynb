## 二項分布

```python
def BinomialDistribution(n, p, x):
    return math.factorial(n) * pow(p,x) * pow(1-p,n-x) /(math.factorial(n-x)*math.factorial(x))   
```

```python
BinomialDistribution(5, 0.3, 0)
fix, axes = plt.subplots(3,3,figsize=(10,13));
n = 10
p = 0.3


for idx1,n in enumerate([10, 200, 300]):
    for idx2p in enumerate([0.2, 0.5, 0.8]):
        x = [a for a in range(0,n+1)]
        y = [BinomialDistribution(n,p,c) for c in x]
        axes[0].plot(x,y)
```
