# データ数が多い場合の多項式分析

```python
from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures(degree = 3)
poly.fit(X.reshape(-1,1))
X_poly_3 = poly.transform(X.reshape(-1,1))



three_model = LinearRegression()
three_model.fit(X_poly_3, y)

X_line=np.linspace(0,1.5,1000)
plt.ylim(-1.5, 1.5)
X_line_poly_3=poly.fit_transform(X_line.reshape(-1,1))
# plot_sin()
plt.plot(np.linspace(0,1,1000),sin_X,"red")
plt.plot(X_line,three_model.predict(X_line_poly_3), color = "green")
```
