## カイ２乗分布

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# X
x = np.linspace(-1, 8, 10000)

# 自由度 k=1〜5
k_deg = [1, 2, 3, 4, 5]

# 自由度 n=1〜5 までのグラフを描画
for i in k_deg:
    plt.plot(x, chi2.pdf(x, i), linestyle='--', label='free k={}'.format(i))

"""
****************************************
グラフの書式設定
****************************************
"""
plt.title("χ2分布")   # タイトル
plt.xlim(-1, 8)       # X軸範囲
plt.ylim(0, 1)       # Y軸範囲
plt.legend()         # 凡例
plt.show()
```
