## カイ２乗分布

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import random
import seaborn as sns
from scipy import stats
```

```python
# X
x = np.linspace(-1, 8, 10000)

# 自由度 k=1〜5
k_deg = [9]

# 自由度 n=1〜5 までのグラフを描画
for i in k_deg:
    plt.plot(x, chi2.pdf(x, i), linestyle='--', label='free k={}'.format(i))

"""
****************************************
グラフの書式設定
****************************************
"""
plt.title("χ2")   # タイトル
plt.xlim(0, 8)       # X軸範囲
plt.ylim(0, 1)       # Y軸範囲
plt.legend()         # 凡例
plt.show()
```

```python
trial = 500000
ret_list = []
for i in range(trial):
    n = 6
    us = np.random.normal(size= n)
    mean_us = np.mean(us)
    std_us = (us - mean_us)@(us - mean_us).T
    ret_list.append(std_us)
```

```python
sns.histplot(ret_list, stat = 'probability',binwidth=0.5, binrange=(0,8))
```

## T分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#-3から3の範囲で100等分
x = np.linspace(-3, 3, 100)
for df in [1,5]:
    t = stats.t.pdf(x, df)
    plt.plot(x, t, label=f"df={df}")


#標準正規分布作成
norm = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, norm, label="norm")
plt.legend()
```

```python
trial = 50000
ret_list = []
for i in range(trial):
    n = 3
    us = np.random.normal(size= n)
    mean_us = np.mean(us)    
    std_us = (us - mean_us)@(us - mean_us).T/n
    ret_list.append(mean_us/np.sqrt(std_us))
```

```python
sns.histplot(ret_list, stat = 'probability',binwidth=0.1, binrange=(-3,3))
```

## 計測

```python
trial = 50000
ret_list = []
for i in range(trial):
    n1 = 100  
    us1 = np.random.normal(size= n1)
    mean_us1 = np.mean(us1)    
    std_us1 = (us1 - mean_us1)@(us1 - mean_us1).T/n1
    
    n2 = 100
    us2 = np.random.normal(size= n2)
    mean_us2 = np.mean(us2)    
    std_us2 = (us2 - mean_us2)@(us2 - mean_us2).T/n2
        
    ret_list.append(std_us1/std_us2)
```

```python
sns.histplot(ret_list,binrange=(0,3),stat='probability')
```
