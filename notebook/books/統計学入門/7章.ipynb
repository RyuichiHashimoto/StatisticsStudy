```python
import matplotlib.pyplot as plt
```

```python
e1 = 0.198
sig1 = 0.357
e2 = 0.055
sig2 = 0.203
ro=0.0
```

```python
def Exp_RP(x):
    return e1*x + e2*(1-x)

def vip_RP(x):
    return (x*sig1)**2 + ((1-x)*sig2)**2 + 2*x*(1-x)*ro*sig1*sig2
    
    
```

```python
xs = [i/100 for i in range(0,100,1)]
es = [ Exp_RP(i) for i in xs]
vs = [ vip_RP(i) for i in xs]
```

```python
plt.plot(xs,es)
```

```python
plt.plot(xs,vs)
```
