```python
import numpy as np
import latexify
import math
import matplotlib.pyplot as plt
from matplotlib import cm
```

## 二項分布

```python
def BinomialDistribution(n, p, x):
    return math.factorial(n)/(math.factorial(n-x))/math.factorial(x) * pow(p,x) * pow(1-p,n-x) 
```

```python
BinomialDistribution(5, 0.3, 0)
fix, axes = plt.subplots(3,3,figsize=(10,13));
n = 10
p = 0.3


for idx1,n in enumerate([10, 20, 30]):
    for idx2,p in enumerate([0.2, 0.5, 0.8]):
        x = [a for a in range(0,n+1)]
        y = [BinomialDistribution(n,p,c) for c in x]
        axes[idx1,idx2].plot(x,y)
        axes[idx1,idx2].set_title(f"n = {n}, p = {p}")        
```

```python
def PoissonDistribution(lmbd, x):
    return math.exp(-1*lmbd)*pow(lmbd, x)/math.factorial(x)
```

```python
fix, axes = plt.subplots(1,3,figsize=(30,6));

for idx1,lambd in enumerate([1, 10, 20]):
    x = [a for a in range(0,30)]
    y = [PoissonDistribution(lambd,c) for c in x]
    axes[idx1].plot(x,y)
    axes[idx1].set_title(f"lambda = {lambd}")
```

## 正規分布

```python
def NormalDistribution(myu, ro, x):
    return 1/math.sqrt(2*math.pi*ro*ro)*math.exp(-1*(x-myu)*(x-myu)/(2*ro*ro))
```

```python
fix, axes = plt.subplots(3,3,figsize=(10,13));
n = 10
p = 0.3


for idx1,myu in enumerate([-30, 0, 30]):
    for idx2, dis in enumerate([1, 10, 100]):
        x = [a/100 for a in range(-30000,30000)]
        y = [NormalDistribution(myu,dis,c) for c in x]
        axes[idx1,idx2].plot(x,y)
        axes[idx1,idx2].set_title(f"myu = {n}, dis = {p}")        
```

```python
## 同時確率分布
x = y = np.arange(-20, 20, 0.5)
X, Y = np.meshgrid(x,y)
z = np.c_[X.ravel(),Y.ravel()]
```

```python
def gaussian(x, sigma, mu):
    #分散共分散行列の行列式
    det = np.linalg.det(sigma)
    #分散共分散行列の逆行列
    inv = np.linalg.inv(sigma)
    n = x.ndim    
    return np.exp(-np.diag((x - mu)@inv@(x - mu).T)/2.0) / (np.sqrt((2 * np.pi) ** n * det))
```

```python
# fix, axes = plt.subplots(3,3,figsize=(10,13));
# #二次元正規分布をplot
idc = 0
for idx1, myu1 in enumerate([-0.8, 0, 0.8]):
        idc += 1
        sig = [[10,myu1*20],[myu1,10]]
        mu = [0,0]
        Z = gaussian(z,sig,mu);
        shape = X.shape
        Z = Z.reshape(shape)        
        fig = plt.figure(figsize = (20, 20))
        ax = fig.add_subplot(330 + idc, projection='3d')
        ax.set_xlim(-20,20)
        ax.set_ylim(-20,20)
        ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
        ax.set_title(f"x_ {myu1}")
            
            


plt.show()
```
